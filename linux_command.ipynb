# ls 명령어 설명
ls 명령어는 유닉스 계열 운영 체제(Linux 등)에서 현재 디렉토리 또는 지정한 디렉토리 내의 파일과 디렉토리 목록을 표시하는 필수 명령어입니다. list segments의 약자로, 파일명, 디렉토리 구조, 그리고 -l 옵션을 통한 권한, 소유자, 크기 등 상세 정보를 확인하는 데 사용됩니다.

# 자주 사용하는 ls 옵션
*   ls -l: 파일의 상세 정보(권한, 소유자, 크기, 수정 시간)를 나열합니다.
*   ls -a: 숨김 파일(점.으로 시작하는 파일)을 포함한 모든 파일을 표시합니다.
*   ls -al: 상세 정보와 함께 숨김 파일을 모두 보여줍니다.
*   ls -h: -l 옵션과 함께 사용하여 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시합니다.
*   ls -t: 파일을 수정 시간 순서대로 최신순으로 정렬하여 표시합니다.
*   ls -r: 목록을 역순으로 정렬합니다.
*   ls -R: 하위 디렉토리의 내용까지 재귀적으로 모두 나열합니다.
*   ls -d: 디렉토리 자체의 정보를 보여줍니다.

# 활용 예시
ls -alF: 모든 파일을 상세 정보와 함께 표시하고, 파일 형식을 알리는 기호(*, / 등)를 끝에 붙입니다.

ls -ltR: 하위 디렉토리를 포함하여 시간순으로 정렬된 목록을 상세히 출력합니다.
기본적으로 ls는 현재 디렉토리의 내용을 보여주며, man ls 명령을 통해 더 자세한 도움말을 확인할 수 있습니다.

Linux 및 유닉스 계열에서 touch 명령어는 0바이트 빈 파일을 생성하거나, 기존 파일의 수정 시간(timestamp)을 현재 시간으로 갱신하는 데 사용되는 기본 명령어입니다. 파일이 없으면 새로 만들고, 있으면 시간 정보만 업데이트하여 관리 효율성을 높입니다.

# 주요 기능 및 사용법:
파일 생성 (가장 빈번한 사용): touch filename - filename이라는 빈 파일을 생성합니다.

시간 갱신 (Update): touch existing_file - 이미 존재하는 파일의 접근 시간(atime)과 수정 시간(mtime)을 현재 시스템 시간으로 변경합니다.

여러 파일 동시 생성: touch file1 file2 file3 - 여러 개의 빈 파일을 한 번에 만듭니다.

# 주요 옵션:

-c: 파일이 없어도 새로 생성하지 않습니다.
-t [YYYYMMDDhhmm]: 현재 시간이 아닌 지정한 시간으로 타임스탬프를 설정합니다.

-r : 디렉토리 내부의 모든 내용을 삭제합니다.

-v : 삭제되는 대상의 정보를 출력합니다.



# cp 명령어 설명
cp (copy) 명령어는 리눅스/유닉스 환경에서 파일이나 디렉토리를 복사하는 데 사용됩니다. 원본 파일은 그대로 유지하며, 이름이나 위치를 바꾸어 새로운 복사본을 생성할 수 있습니다. 기본 구조는 cp [옵션] <원본> <대상>입니다.

# 주요 활용법 및 옵션:
파일 복사: cp file1.txt file1_backup.txt (file1.txt를 file1_backup.txt로 복사)

디렉토리 복사: cp -r dir1 dir2 (디렉토리 및 하위 내용까지 모두 복사)

복사 시 묻지 않고 덮어쓰기: \cp -f source target (alias 설정 무시하고 강제
덮어쓰기)

파일 속성 유지 복사: cp -p source target (원본 파일의 권한, 시간 등 정보 유지)

복사 과정 확인: cp -v source target (복사 진행 상황 화면 출력)

# mv 명령어 설명
mv (move) 명령어는 리눅스 및 유닉스 시스템에서 파일이나 디렉토리를 다른 위치로 이동하거나, 이름을 변경할 때 사용하는 핵심 명령어입니다. 기본적으로 mv [옵션] [원본] [대상] 형식을 사용하며, 원본 파일은 삭제되고 대상 위치에 새로운 파일로 저장됩니다.

파일/디렉토리 이동: mv file1.txt /home/user/documents/ (file1.txt를 documents 폴더로 이동).

파일/디렉토리 이름 변경: mv old_name.txt new_name.txt (파일 이름을 old에서 new로 변경).

여러 파일 이동: mv file1.txt file2.txt /folder/ (여러 파일을 한 번에 폴더로 이동).

# 자주 사용하는 옵션
-i (interactive): 이동할 위치에 동일한 이름의 파일이 있을 경우, 덮어쓸지 물어봅니다.

-n (no-clobber): 기존 파일을 덮어쓰지 않습니다.

-v (verbose): 이동 상황을 화면에 표시합니다.


# 주의사항
대상 위치에 동일한 이름의 파일이 있으면 별도 옵션 없이 mv 명령어를 실행할 경우, 원본 파일이 덮어쓰기 되어 복구할 수 없게 되므로 주의해야 합니다.

# mkdir 명령어 설명
mkdir(make directory)은 리눅스, 유닉스, macOS 등 CLI 환경에서 디렉토리(폴더)를 생성하는 명령어입니다. 기본적으로 mkdir [폴더명] 형태로 사용하며, -p 옵션을 통해 상위 디렉토리까지 함께 생성하거나 -m 옵션으로 권한을 설정할 수 있습니다.

# 핵심 사용법
기본 생성: mkdir folder_name (현재 위치에 folder_name 생성)

여러 폴더 생성: mkdir dir1 dir2 dir3 (공백으로 구분)

상위 폴더 포함 생성 (-p): mkdir -p A/B/C (상위 디렉토리가 없으면 함께 생성)

권한 설정 생성 (-m): mkdir -m 755 new_folder (생성 시 퍼미션(권한) 지정)

생성 내용 출력 (-v): mkdir -v new_folder (디렉토리 생성 시 메시지 출력)


# 주요 특징
이미 존재하는 폴더명으로 생성 시 에러가 발생하나, -p 옵션을 사용하면 에러 없이 건너뜁니다.

# cat 명령어 설명
cat (concatenate) 명령어는 리눅스/유닉스 환경에서 파일 내용을 화면에 출력하거나, 파일을 합치고 생성하는 데 사용되는 가장 기초적이고 강력한 도구입니다. 파일 확인, 파일 내용 연속 출력, 리다이렉션(>)을 이용한 새 파일 생성 및 내용 추가가 주된 기능입니다.

# 주요 활용법 및 예제
파일 내용 확인: cat filename (파일의 전체 내용 출력).

여러 파일 합치기: cat file1 file2 > combined.txt (file1, file2를 하나로 합쳐 combined.txt 생성).

파일 생성 (간단한 내용): cat > newfile.txt (내용 입력 후 Ctrl+D로 저장 및 종료).

줄 번호 포함 출력: cat -n filename (빈 줄 포함하여 줄 번호 표시).

파일 끝 표시: cat -E filename (각 줄의 끝에 $ 문자를 표시).

탭(Tab) 문자를 ^I로 출력: cat -T filename.

# vim 명령어 설명
Vim은 명령 모드(ESC)에서 h, j, k, l로 이동, i로 입력 모드 전환, :wq로 저장 후 종료하는 강력한 텍스트 에디터입니다. :q!로 강제 종료, x로 글자 삭제, yy로 줄 복사, p로 붙여넣기가 가능하며, 명령어를 조합해 빠른 편집을 수행합니다.

# 핵심 Vim 명령어 요약
# 모드 전환
ESC: 명령 모드 (기본 상태)

i: 현재 커서 위치에서 입력 모드

a: 현재 커서 다음 위치에서 입력 모드

o: 현재 줄 아래에 새로운 줄을 열고 입력 모드

# 저장 및 종료 (명령 모드에서 : 입력 후)
:w: 저장

:q: 종료

:wq 또는 ZZ: 저장 후 종료

:q!: 저장하지 않고 강제 종료

# 이동 (명령 모드)
h, j, k, l: 왼쪽, 아래, 위, 오른쪽

0: 줄의 맨 앞

$: 줄의 맨 뒤

gg: 문서 맨 처음

G: 문서 맨 끝

dw: 단어 삭제

dd: 현재 줄 삭제

u: 실행 취소 (Undo)

yy: 현재 줄 복사 (Yank)

p: 붙여넣기 (Paste)

# chmod 명령어 설명
chmod(change mode)는 유닉스 및 리눅스 계열 운영체제에서 파일이나 디렉토리의 접근 권한(Permission)을 변경하는 명령어입니다. 사용자(u), 그룹(g), 기타 사용자(o)에 대해 읽기(r, 4), 쓰기(w, 2), 실행(x, 1) 권한을 8진수 숫자나 심볼릭 기호로 설정하여 보안을 관리합니다.

# 주요 특징 및 사용법
기본 구조: chmod [옵션] [권한] [파일/디렉토리]

8진수 모드: 숫자의 합으로 권한 설정

심볼릭 모드: 문자를 사용하여 권한 수정 (예: u+x는 사용자에게 실행 권한 추가)

재귀적 변경: -R 옵션을 사용하여 디렉토리 하위 파일까지 한 번에 권한 변경 가능

# 자주 사용하는 명령어 예시
chmod 755 filename: 소유자(모든 권한), 그룹/기타(읽기+실행)

chmod 644 filename: 소유자(읽기+쓰기), 그룹/기타(읽기)

chmod 777 filename: 모든 사용자에게 모든 권한 부여 (보안 위험)

chmod +x script.sh: 스크립트에 실행 권한 추

# chown 명령어 설명
hown (change owner)은 리눅스 및 유닉스 시스템에서 파일이나 디렉토리의 소유자(User) 및 그룹(Group)을 변경하는 명령어입니다. 보통 관리자 권한(root)이 필요하며, -R 옵션을 통해 하위 파일 및 디렉토리의 소유권까지 일괄 변경할 수 있습니다.

# 주요 명령어 구조:
chown [옵션] [소유자]:[소유그룹] [파일/디렉토리]

사용 예시:

소유자만 변경: chown user1 file.txt

소유자와 그룹 변경: chown user1:group1 file.txt

하위 디렉토리까지 일괄 변경 (Recursive): chown -R user1:group1 /path/to/dir

# 핵심 특징:
권한: 파일 소유자를 변경하려면 root 권한이 필요합니다.

chgrp과 차이: chown은 소유자와 그룹 모두, chgrp은 그룹만 변경합니다.

-R 옵션: 디렉토리 내부의 모든 파일과 서브 디렉토리의 소유권을 변경합니다.

# 자주 사용하는 옵션:
-R: 하위 디렉토리와 파일까지 모두 변경 (Recursive)

-v: 변경된 내용을 자세히 출력 (Verbose)

-h: 심볼릭 링크 자체의 소유권을 변경

참고: 권한은 chmod 명령어로 변경하고, 소유권은 chown으로 변경합니다.

# chgrp 명령어 설명
chgrp(change group)는 유닉스 및 리눅스 계열 운영체제에서 파일이나 디렉토리의 소유 그룹을 변경하는 명령어입니다. 주로 시스템 관리자가 파일에 대한 접근 권한을 그룹 단위로 관리하기 위해 사용하며, 기본 형식은 chgrp [옵션] [그룹] [파일/디렉토리]입니다.

# 핵심 사용법 및 옵션
기본 변경: chgrp newgroup file.txt (file.txt의 그룹을 newgroup으로 변경)

하위 디렉토리 포함 변경 (-R): chgrp -R newgroup directory/ (디렉토리와 하위 항목 모두 변경)

심볼릭 링크 변경 (-h): 링크 자체가 가리키는 파일이 아닌 링크 파일 자신의 그룹을 변경.

# 주요 특징
일반 사용자는 자신이 속한 그룹으로만 변경 가능하며, root 사용자는 모든 그룹으로 변경할 수 있습니다.

그룹 이름이나 숫자 그룹 ID(GID)를 사용할 수 있습니다.

유사 명령어로 소유자와 그룹을 동시에 변경하는 chown이 있습니다.
