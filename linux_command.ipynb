# ls 명령어 설명
ls 명령어는 유닉스 계열 운영 체제(Linux 등)에서 현재 디렉토리 또는 지정한 디렉토리 내의 파일과 디렉토리 목록을 표시하는 필수 명령어입니다. list segments의 약자로, 파일명, 디렉토리 구조, 그리고 -l 옵션을 통한 권한, 소유자, 크기 등 상세 정보를 확인하는 데 사용됩니다.

# 자주 사용하는 ls 옵션
*   ls -l: 파일의 상세 정보(권한, 소유자, 크기, 수정 시간)를 나열합니다.
*   ls -a: 숨김 파일(점.으로 시작하는 파일)을 포함한 모든 파일을 표시합니다.
*   ls -al: 상세 정보와 함께 숨김 파일을 모두 보여줍니다.
*   ls -h: -l 옵션과 함께 사용하여 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시합니다.
*   ls -t: 파일을 수정 시간 순서대로 최신순으로 정렬하여 표시합니다.
*   ls -r: 목록을 역순으로 정렬합니다.
*   ls -R: 하위 디렉토리의 내용까지 재귀적으로 모두 나열합니다.
*   ls -d: 디렉토리 자체의 정보를 보여줍니다.

# 활용 예시
ls -alF: 모든 파일을 상세 정보와 함께 표시하고, 파일 형식을 알리는 기호(*, / 등)를 끝에 붙입니다.

ls -ltR: 하위 디렉토리를 포함하여 시간순으로 정렬된 목록을 상세히 출력합니다.
기본적으로 ls는 현재 디렉토리의 내용을 보여주며, man ls 명령을 통해 더 자세한 도움말을 확인할 수 있습니다.

# touch 명령어 설명
Linux 및 유닉스 계열에서 touch 명령어는 0바이트 빈 파일을 생성하거나, 기존 파일의 수정 시간(timestamp)을 현재 시간으로 갱신하는 데 사용되는 기본 명령어입니다. 파일이 없으면 새로 만들고, 있으면 시간 정보만 업데이트하여 관리 효율성을 높입니다.

# 주요 기능 및 사용법:
파일 생성 (가장 빈번한 사용): touch filename - filename이라는 빈 파일을 생성합니다.

시간 갱신 (Update): touch existing_file - 이미 존재하는 파일의 접근 시간(atime)과 수정 시간(mtime)을 현재 시스템 시간으로 변경합니다.

여러 파일 동시 생성: touch file1 file2 file3 - 여러 개의 빈 파일을 한 번에 만듭니다.

# 주요 옵션:
-a: 접근 시간(access time)만 현재 시간으로 변경합니다.

-m: 수정 시간(modification time)만 현재 시간으로 변경합니다.

-c: 파일이 없어도 새로 생성하지 않습니다.
-t [YYYYMMDDhhmm]: 현재 시간이 아닌 지정한 시간으로 타임스탬프를 설정합니다.

주로 스크립트 작성 시 빈 파일을 만들거나, 빌드 시스템에서 파일의 최신 버전을 표시할 때 활용됩니다.

-r : 디렉토리 내부의 모든 내용을 삭제합니다.

-v : 삭제되는 대상의 정보를 출력합니다.



# cp 명령어 설명
cp (copy) 명령어는 리눅스/유닉스 환경에서 파일이나 디렉토리를 복사하는 데 사용됩니다. 원본 파일은 그대로 유지하며, 이름이나 위치를 바꾸어 새로운 복사본을 생성할 수 있습니다. 기본 구조는 cp [옵션] <원본> <대상>입니다.

# 주요 활용법 및 옵션:
파일 복사: cp file1.txt file1_backup.txt (file1.txt를 file1_backup.txt로 복사)

디렉토리 복사: cp -r dir1 dir2 (디렉토리 및 하위 내용까지 모두 복사)

복사 시 묻지 않고 덮어쓰기: \cp -f source target (alias 설정 무시하고 강제
덮어쓰기)

파일 속성 유지 복사: cp -p source target (원본 파일의 권한, 시간 등 정보 유지)

복사 과정 확인: cp -v source target (복사 진행 상황 화면 출력)

# mv 명령어 설명
mv (move) 명령어는 리눅스 및 유닉스 시스템에서 파일이나 디렉토리를 다른 위치로 이동하거나, 이름을 변경할 때 사용하는 핵심 명령어입니다. 기본적으로 mv [옵션] [원본] [대상] 형식을 사용하며, 원본 파일은 삭제되고 대상 위치에 새로운 파일로 저장됩니다.

파일/디렉토리 이동: mv file1.txt /home/user/documents/ (file1.txt를 documents 폴더로 이동).

파일/디렉토리 이름 변경: mv old_name.txt new_name.txt (파일 이름을 old에서 new로 변경).

여러 파일 이동: mv file1.txt file2.txt /folder/ (여러 파일을 한 번에 폴더로 이동).

# 자주 사용하는 옵션
-i (interactive): 이동할 위치에 동일한 이름의 파일이 있을 경우, 덮어쓸지 물어봅니다.

-n (no-clobber): 기존 파일을 덮어쓰지 않습니다.

-v (verbose): 이동 상황을 화면에 표시합니다.


# 주의사항
대상 위치에 동일한 이름의 파일이 있으면 별도 옵션 없이 mv 명령어를 실행할 경우, 원본 파일이 덮어쓰기 되어 복구할 수 없게 되므로 주의해야 합니다.

# mkdir 명령어 설명
mkdir(make directory)은 리눅스, 유닉스, macOS 등 CLI 환경에서 디렉토리(폴더)를 생성하는 명령어입니다. 기본적으로 mkdir [폴더명] 형태로 사용하며, -p 옵션을 통해 상위 디렉토리까지 함께 생성하거나 -m 옵션으로 권한을 설정할 수 있습니다.

# 핵심 사용법
기본 생성: mkdir folder_name (현재 위치에 folder_name 생성)

여러 폴더 생성: mkdir dir1 dir2 dir3 (공백으로 구분)

상위 폴더 포함 생성 (-p): mkdir -p A/B/C (상위 디렉토리가 없으면 함께 생성)

권한 설정 생성 (-m): mkdir -m 755 new_folder (생성 시 퍼미션(권한) 지정)

생성 내용 출력 (-v): mkdir -v new_folder (디렉토리 생성 시 메시지 출력)


# 주요 특징
이미 존재하는 폴더명으로 생성 시 에러가 발생하나, -p 옵션을 사용하면 에러 없이 건너뜁니다.

# cat 명령어 설명
cat (concatenate) 명령어는 리눅스/유닉스 환경에서 파일 내용을 화면에 출력하거나, 파일을 합치고 생성하는 데 사용되는 가장 기초적이고 강력한 도구입니다. 파일 확인, 파일 내용 연속 출력, 리다이렉션(>)을 이용한 새 파일 생성 및 내용 추가가 주된 기능입니다.

# 주요 활용법 및 예제
파일 내용 확인: cat filename (파일의 전체 내용 출력).

여러 파일 합치기: cat file1 file2 > combined.txt (file1, file2를 하나로 합쳐 combined.txt 생성).

파일 생성 (간단한 내용): cat > newfile.txt (내용 입력 후 Ctrl+D로 저장 및 종료).

줄 번호 포함 출력: cat -n filename (빈 줄 포함하여 줄 번호 표시).

파일 끝 표시: cat -E filename (각 줄의 끝에 $ 문자를 표시).

탭(Tab) 문자를 ^I로 출력: cat -T filename.

# chmod 명령어 설명
chmod(change mode)는 유닉스 및 리눅스 계열 운영체제에서 파일이나 디렉토리의 접근 권한(Permission)을 변경하는 명령어입니다. 사용자(u), 그룹(g), 기타 사용자(o)에 대해 읽기(r, 4), 쓰기(w, 2), 실행(x, 1) 권한을 8진수 숫자나 심볼릭 기호로 설정하여 보안을 관리합니다.

# 주요 특징 및 사용법
기본 구조: chmod [옵션] [권한] [파일/디렉토리]

8진수 모드: 숫자의 합으로 권한 설정

심볼릭 모드: 문자를 사용하여 권한 수정 (예: u+x는 사용자에게 실행 권한 추가)

재귀적 변경: -R 옵션을 사용하여 디렉토리 하위 파일까지 한 번에 권한 변경 가능

# 자주 사용하는 명령어 예시
chmod 755 filename: 소유자(모든 권한), 그룹/기타(읽기+실행)

chmod 644 filename: 소유자(읽기+쓰기), 그룹/기타(읽기)

chmod 777 filename: 모든 사용자에게 모든 권한 부여 (보안 위험)

chmod +x script.sh: 스크립트에 실행 권한 추

# chown 명령어 설명
hown (change owner)은 리눅스 및 유닉스 시스템에서 파일이나 디렉토리의 소유자(User) 및 그룹(Group)을 변경하는 명령어입니다. 보통 관리자 권한(root)이 필요하며, -R 옵션을 통해 하위 파일 및 디렉토리의 소유권까지 일괄 변경할 수 있습니다.

# 주요 명령어 구조:
chown [옵션] [소유자]:[소유그룹] [파일/디렉토리]

사용 예시:

소유자만 변경: chown user1 file.txt

소유자와 그룹 변경: chown user1:group1 file.txt

하위 디렉토리까지 일괄 변경 (Recursive): chown -R user1:group1 /path/to/dir

# 핵심 특징:
권한: 파일 소유자를 변경하려면 root 권한이 필요합니다.

chgrp과 차이: chown은 소유자와 그룹 모두, chgrp은 그룹만 변경합니다.

-R 옵션: 디렉토리 내부의 모든 파일과 서브 디렉토리의 소유권을 변경합니다.

# 자주 사용하는 옵션:
-R: 하위 디렉토리와 파일까지 모두 변경 (Recursive)

-v: 변경된 내용을 자세히 출력 (Verbose)

-h: 심볼릭 링크 자체의 소유권을 변경

참고: 권한은 chmod 명령어로 변경하고, 소유권은 chown으로 변경합니다.

# chgrp 명령어 설명
chgrp(change group)는 유닉스 및 리눅스 계열 운영체제에서 파일이나 디렉토리의 소유 그룹을 변경하는 명령어입니다. 주로 시스템 관리자가 파일에 대한 접근 권한을 그룹 단위로 관리하기 위해 사용하며, 기본 형식은 chgrp [옵션] [그룹] [파일/디렉토리]입니다.

# 핵심 사용법 및 옵션
기본 변경: chgrp newgroup file.txt (file.txt의 그룹을 newgroup으로 변경)

하위 디렉토리 포함 변경 (-R): chgrp -R newgroup directory/ (디렉토리와 하위 항목 모두 변경)

심볼릭 링크 변경 (-h): 링크 자체가 가리키는 파일이 아닌 링크 파일 자신의 그룹을 변경.

# 주요 특징
일반 사용자는 자신이 속한 그룹으로만 변경 가능하며, root 사용자는 모든 그룹으로 변경할 수 있습니다.

그룹 이름이나 숫자 그룹 ID(GID)를 사용할 수 있습니다.

유사 명령어로 소유자와 그룹을 동시에 변경하는 chown이 있습니다.

# apt 명령어 설명
apt는 우분투, 데비안 등 리눅스에서 패키지를 설치, 업데이트, 삭제하는 강력한 명령줄 도구입니다. 주로 sudo 권한이 필요하며, 주요 명령어로 update(목록 갱신), upgrade(패키지 업그레이드), install(설치), remove(삭제)를 사용하여 의존성 문제를 자동으로 해결합니다.

sudo apt autoremove: 사용하지 않거나 의존성이 해제된 패키지 자동 삭제.

# tar 명령어 설명
tar 명령은 리눅스/유닉스 환경에서 여러 파일이나 디렉토리를 하나의 .tar 파일(아카이브)로 묶거나, 묶인 파일을 해제하는 도구입니다. gzip 등과 결합하여 압축(tar.gz)도 가능하며, 주요 옵션은 묶기(-c), 풀기(-x), 상세정보 표시(-v), 파일 지정(-f)입니다.

# 자주 사용하는 tar 명령어 패턴
파일/폴더 묶기 (아카이브 생성)
tar -cvf [생성할파일.tar] [대상폴더 또는 파일]

tar 파일 풀기
tar -xvf [파일.tar]

tar.gz 압축하기
tar -zcvf [생성할파일.tar.gz] [대상폴더 또는 파일]

tar.gz 압축 풀기
tar -zxvf [파일.tar.gz]

tar 파일 내용 확인
tar -tvf [파일.tar]

# 주요 옵션 설명
-c: Create, 새로운 묶음 파일 생성

-x: Extract, 묶음 파일에서 파일 추출(압축 해제)

-v: Verbose, 작업 과정을 화면에 상세히 표시

-f: File, 아카이브 파일명을 지정 (필수 옵션)

-z: gzip(compress) 방식으로 압축 또는 해제

-j: bzip2 방식으로 압축 또는 해제

# 팁
tar는 파일들을 하나로 묶기만 하고, 압축 기능은 별도 옵션(-z, -j)을 사용해야 합니다.

-f 옵션 뒤에는 반드시 아카이브 파일명(tar 파일)이 위치해야 합니다.

systemctl은 리눅스에서 systemd 기반의 서비스와 데몬을 관리하는 핵심 명령어로, 서비스의 시작, 중지, 상태 확인, 활성화/비활성화 등을 수행합니다. systemctl [명령어] [서비스명] 형식을 사용하며, 주로 start, stop, restart, status, enable, disable 명령어를 활용하여 시스템을 제어합니다.

# nslookup 명령어 설명
nslookup (Name Server Lookup)은 도메인 네임 시스템(DNS)에 쿼리하여 도메인 이름에 대한 IP 주소 매핑이나 레코드 정보를 조회하는 명령줄 도구입니다. Windows, Linux, macOS 등 대부분의 운영체제에서 기본적으로 사용 가능하며, 네트워크 진단, DNS 서버 상태 확인 및 설정 변경 사항을 확인하는 데 주로 사용됩니다.

# 주요 명령어 및 활용법:
기본 조회: nslookup [도메인명] (예: nslookup google.com) - 해당 도메인의 IP 주소(A 레코드) 및 DNS 서버 정보를 확인합니다.

특정 DNS 서버 조회: nslookup [도메인명] [DNS서버IP] - 특정 DNS 서버(예: 8.8.8.8)를 지정하여 해당 도메인의 정보를 조회합니다.

레코드 타입 지정: nslookup -type=[유형] [도메인명] (예: nslookup -type=mx google.com) - MX, NS, SOA 등 특정 DNS 레코드를 조회할 때 사용합니다.

대화형 모드: nslookup 입력 후 실행하여, 지속적으로 DNS 쿼리를 날릴 수 있는 대화형 상태로 전환할 수 있습니다.

# 결과 해석:

# pipe(|) 명령어 설명
리눅스 및 유닉스 계열에서 파이프(|)는 앞선 명령어의 표준 출력(stdout)을 다음 명령어의 표준 입력(stdin)으로 직접 전달하는 핵심 기능입니다. 여러 명령어를 연결하여 복잡한 데이터 처리를 자동화하는 파이프라인을 구축할 때 사용되며, 일방향 데이터 흐름을 제공합니다.

# redirect 명령어 설명
리눅스/유닉스 리디렉션(>,>>,<)은 명령어의 표준 입출력 방향을 화면에서 파일이나 장치로 변경하는 기능입니다. 출력 저장(>), 추가(>>), 입력 받기(<)를 통해 결과 기록, 로그 파일 생성, 스크립트 자동화에 필수적입니다.

# 주요 리디렉션 명령어 및 기호
> 파일명 (출력 재지향 - 덮어쓰기): 명령의 결과를 파일로 저장하며, 기존 파일 내용은 삭제됩니다.

예: ls -l > file_list.txt (명령 결과를 file_list.txt에 저장)

>> 파일명 (출력 재지향 - 이어쓰기): 명령의 결과를 파일 끝에 추가합니다.

예: echo "new log" >> system.log (로그 파일에 내용 추가)

< 파일명 (입력 재지향): 명령어가 키보드가 아닌 파일로부터 입력을 받도록 합니다.

예: sort < input.txt (input.txt 내용을 정렬)

2> 파일명 (에러 재지향): 에러 메시지만 파일로 따로 저장합니다.

예: command 2> error.log

> 파일명 2>&1 (전체 출력 재지향): 표준 출력과 에러를 모두 한 파일로 보냅니다.

예: command > output.log 2>&1

/dev/null (출력 버리기): 불필요한 출력이나 에러를 '블랙홀' 같은 장치로 보내 삭제합니다.

예: command > /dev/null 2>&1 (결과 및 에러 모두 화면에 표시하지 않음)

# 핵심 팁:
 >  >를 잘못 사용하면 기존 파일을 덮어쓰므로 주의해야 합니다.

# ps 명령어 설명
ps(Process Status) 명령어는 리눅스/유닉스 시스템에서 현재 실행 중인 프로세스의 상태를 스냅샷 형태로 확인하는 도구입니다. 주요 옵션으로 모든 프로세스를 상세히 출력하는 ps -ef나 ps aux를 사용하여 프로세스의 ID(PID), CPU/메모리 사용량, 실행 명령 등을 파악해 시스템 모니터링 및 프로세스 관리에 활용합니다.

# 주요 ps 명령어 옵션 및 예제
ps -ef: 시스템의 모든 프로세스를 자세한 형식(PPID 포함)으로 출력.

ps aux: BSD 스타일로 모든 프로세스의 상세 정보(CPU/메모리 사용률 포함)를 표시.

ps -u [사용자이름]: 특정 사용자의 프로세스만 조회.

ps -p [PID]: 특정 프로세스 ID에 대한 정보만 확인.

ps -eo pid,ppid,cmd,%mem,%cpu --sort=-%mem: 메모리 사용량이 높은 순으로 프로세스 정렬하여 출력.

UID/USER: 프로세스 소유자.

PID: 프로세스 고유 번호.

PPID: 부모 프로세스 PID.

C/CPU: CPU 사용률.

STIME/START: 프로세스 시작 시간.

TIME: 프로세스가 사용한 총 CPU 시간.

CMD/COMMAND: 실행된 명령어.

팁: 프로세스를 실시간으로 모니터링하려면 top이나 htop 명령어를 사용하는 것이 좋습니다.

# kill 명령어 설명
리눅스 kill 명령어는 실행 중인 프로세스에 시그널(Signal)을 보내 작업을 종료하거나 제어하는 도구입니다. 주로 응답 없는 프로그램을 종료할 때 사용하며, PID(프로세스 ID)를 기반으로 작동합니다. 정상 종료(SIGTERM, 15)가 기본이며, 필요시 강제 종료(SIGKILL, 9)를 사용합니다.

강제 종료(즉시): kill -9 [PID] (응답 없는 프로세스용)

특정 사용자/이름 종료: pkill -f [프로세스명]

# status 명령어 설명
리눅스에서 서비스의 상태를 확인하는 가장 대표적인 명령어는 systemctl status <서비스명>입니다. 이 명령어는 서비스가 현재 실행 중(active)인지, 멈췄는지(inactive), 그리고 최근 로그를 포함한 상세 정보를 제공합니다. 주요 서비스 관리 명령어는 다음과 같습니다.

# 핵심 서비스 상태 확인 명령어
systemctl status <서비스명>: 특정 서비스(예: nginx, docker)의 상세 상태 확인.

systemctl is-active <서비스명>: 서비스의 active/inactive 여부만 간결하게 확인.

systemctl list-units --type=service: 현재 실행 중인 모든 서비스 목록 확인.

service --status-all: + 및 - 표시로 모든 서비스 상태 확인.

# 활용 예시
systemctl status sshd : SSH 서비스의 현재 상태 및 로그 확인.

systemctl status mysql : MySQL 데이터베이스 상태 확인.

# restart 명령어 설명
리눅스에서 시스템을 즉시 재시작(Restart/Reboot)하는 가장 대표적인 명령어는 sudo reboot 또는 sudo shutdown -r now입니다. 시스템 관리자 권한이 필요하며, 사용 중인 프로세스를 안전하게 종료하고 시스템을 다시 시작합니다.

# 주요 리눅스 재시작 명령어:
sudo reboot: 가장 간편하고 일반적인 즉시 재부팅 명령어.

sudo shutdown -r now: 현재 시간(now)에 시스템을 재시작(-r)하는 명령어.

sudo shutdown -r +10: 10분 후 재시작 예약.

sudo systemctl reboot: systemd 기반 시스템(대부분의 최신 리눅스)에서 사용하는 표준 재부팅 명령어.

init 6: 프로세스를 정석대로 종료하고 시스템을 재시작하는 전통적인 방식.

# 주의사항:
대부분의 명령은 sudo (루트 권한)가 필요합니다.

shutdown -c: 예약된 재시작/종료 명령을 취소합니다.

```python
server.py
import socket

HOST = '0.0.0.0'  # 모든 IP에서 접속 허용
PORT = 12345

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind((HOST, PORT))
server_socket.listen()

print(f"서버가 {PORT} 포트에서 대기 중...")

conn, addr = server_socket.accept()
print(f"{addr} 연결됨")

while True:
    data = conn.recv(1024)
    if not data:
        break
    print("클라이언트:", data.decode())

conn.close()
server_socket.close()
```

```python
client.py
import socket

HOST = '127.0.0.1'  # 서버 IP (같은 컴퓨터면 localhost)
PORT = 12345

client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client_socket.connect((HOST, PORT))

print("서버에 연결됨. 메시지를 입력하세요.")

while True:
    message = input(">>> ")
    if message.lower() == 'exit':
        break
    client_socket.sendall(message.encode())

client_socket.close()
```
